# Task 2

```python
month_dict={
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
def rename_my_file(old_file_name):
    try: 
        month = old_file_name[0:3]
        new_file_name = month_dict[month]+ old_file_name[3:]
        new_file_name = new_file_name.replace('_','-')
        return new_file_name
    except:
        print(f'File has been renamed.')

old_file_name = 'jan-08-singapore.txt'
print(rename_my_file(old_file_name))
```

```python
old_file_names = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']
for filenames in old_file_names:
    print(f'{filenames} --> {rename_my_file(filenames)}')
```

# Task 4

```python
all_old_file_names = glob.glob('Data/*.txt') #filter out all the txt
for filenames in all_old_file_names:
    file_names = os.path.split(filenames)[-1] #isolate filenames from Data
    x = rename_my_file(file_names) #use function to rename filenames
    path = os.path.join('Data', x) #join back Data with new filenames
    os.rename(filenames, path) # (source, dest)
    print(f'Creating {path}')
```
