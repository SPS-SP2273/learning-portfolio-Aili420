# 1 A simple plot

```python
from matplotlib import pyplot as plt
```

## 1.1 Letâ€™s look at some code.

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) # format the grid, smaller no = fainter grid
plt.legend(loc='upper left')
```

```python
plt.plot(x, y_2, label='Y values',
         c='red', ls='dashed',lw=2) # short form
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) # format the grid, smaller no = fainter grid
plt.legend(loc='upper left')
```

```python
plt.plot(x, y_1, y_2) # including y_2 values in graph
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) 
```

## 1.2 Adding another plot

```python
plt.plot(x, y_2,
         label='Y2 values', c='blue', ls='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

```python
# first plot
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) # format the grid, smaller no = fainter grid

# add another plot
plt.plot(x, y_2,
         label='Y2 values', c='blue', ls='solid') # ls is solid by default
plt.legend(loc='upper left') # includes both set of values
```

## 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3,
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green") # constant errors for all points
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left') # put at the end to include all plots
```

# 2 Better with NumPy

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

import numpy as np # convert phython lists into numpy arrays
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values') # solid line

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # math function on x
sin_x = np.sin(x)          # np allows math to be done

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 

plt.xlabel('x-values')
plt.ylabel('math function on x')
plt.title('math function in plot')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6]) # inaccuracy in the plot due to limited number of values
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')   # For y-axis 1

plt.twinx()              # This creates a new y-axis 
                         # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')  # For y-axis 2

plt.xlabel('x-values')
plt.ylabel('math function on x')
plt.title('math function in plot')
plt.grid(alpha=.25)
```

```python
# x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.arange(0, 6, .5) # using np to give a range
print(x)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')   # For y-axis 1

plt.twinx()              # This creates a new y-axis 
                         # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')  # For y-axis 2

plt.xlabel('x-values')
plt.ylabel('math function on x')
plt.title('math function in plot')
plt.grid(alpha=.25)
```

```python
# x = np.array([0, 1, 2, 3, 4, 5, 6])
# x = np.arange(0, 6, .5)
x = np.linspace(0, 6, 100) # using np.linspace()
print(x)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc=(.5, .5))   # For y-axis 1

plt.twinx()              # This creates a new y-axis 
                         # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc=(.5, .6))  # For y-axis 2

plt.xlabel('x-values')
plt.ylabel('math function on x')
plt.title('math function in plot')
plt.grid(alpha=.25)
```

```python
# x = np.array([0, 1, 2, 3, 4, 5, 6])
# x = np.arange(0, 6, .5)
x = np.linspace(0, 6, 100) # using np.linspace()
print(x)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc=(.25, .5), frameon=False)   # For y-axis 1

plt.twinx()              # This creates a new y-axis 
                         # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc=(.25, .6), frameon=False)  # For y-axis 2

plt.xlabel('x-values')
plt.ylabel('math function on x')
plt.title('math function in plot')
plt.grid(alpha=.25)
```

# 3 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.xlabel('x-values')
plt.ylabel('math function on x')
plt.title('math function in plot')
plt.grid(alpha=.25)

plt.savefig('simple-plot.png', dpi=150) # save plot to local drive
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-3, 3])

plt.xlabel('Years')
plt.ylabel('Change in temperature')
plt.title('Average temperature data for Daily Land')
plt.grid(alpha=.25)
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-3, 3])

plt.xlabel('Years')
plt.ylabel('Change in temperature')
plt.title('Average temperature data for Daily Land')
plt.grid(alpha=.25)

# add horizontal line at zero value to highlight trend
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```
